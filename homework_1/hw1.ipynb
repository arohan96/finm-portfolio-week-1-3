```python
# 2.3 The MV frontier

# Compute and display the weights of the tangency portfolios:  w^tan
def return_tangent_weights(returns):
    cov = returns.cov()
    cov_inv = np.linalg.inv(cov*12) # compute the inverse of a square matrix
    mu = returns.mean()*12

    ones = np.ones(returns.columns[1:].shape)
    scaling = 1/(np.transpose(ones) @ cov_inv @ mu)
    tangent_return = scaling*(cov_inv @ mu)
    tangent_weights = pd.DataFrame(index = returns.columns[1:], data = tangent_return, columns = ['Tangent Weights'])
    return tangent_weights

tangent_weights = return_tangent_weights(excess_returns)
tangent_weights
```

```python
# Does the ranking of weights align with the ranking of Sharpe ratios?
compare = tangent_weights.merge(result_1['Sharpe ratio'], on=tangent_weights.index).rename(columns={'key_0':'assets'}).set_index('assets')
print(compare)
print("The ranking of weights does not align with the ranking of Sharpe ratios")
```

```python
# Compute the mean, volatility, and Sharpe ratio for the tangency portfolio corresponding to  w^tan
excess_returns_1 = excess_returns[excess_returns.columns[1:]]
w_t_result = excess_returns_1@tangent_weights
w_t_stats = pd.DataFrame(index=w_t_result.columns)
w_t_stats['mean'] = w_t_result.mean()*12
w_t_stats['volatility'] = w_t_result.std()*np.sqrt(12)
w_t_stats['Sharpe ratio'] = (w_t_stats['mean']/w_t_stats['volatility'])*np.sqrt(12)
w_t_stats
```

```python
# 2.4 TIPS

# Assess how much the tangency portfolio (and performance) change if...
#(1)TIPS are dropped completely from the investment set.
w_t_without_TIPS = return_tangent_weights(excess_returns.drop(columns=['TIP']))
w_t_without_TIPS
```

```python
#(2)The expected excess return to TIPS is adjusted to be 0.0012 higher than what the historic sample shows.
TIP_new = 0.0012
excess_returns_adj = excess_returns.copy()
excess_returns_adj['TIP'] += TIP_new
w_t_adj_TIPS = return_tangent_weights(excess_returns_adj)
w_t_adj_TIPS
```

```python
# 3. Allocations

def return_reg_tangent_weights(returns):
    p = 0.5
    cov = returns.cov()
    cov_mat_diag = np.diag(np.diag(cov))
    cov_mat = p*cov + p*cov_mat_diag
    cov_inv = np.linalg.inv(cov_mat*12) # compute the inverse of a square matrix

    ones = np.ones(returns.columns[1:].shape)
    mu = returns.mean()*12
    scaling = 1/(np.transpose(ones) @ cov_inv @ mu)
    tangent_return = scaling*(cov_inv @ mu)
    tangent_weights = pd.DataFrame(index = returns.columns[1:], data = tangent_return, columns = ['Tangent Weights'])
    return tangent_weights


target_mean = 0.01
excess_returns_3 = excess_returns[excess_returns.columns[1:]]
w_t_compare = pd.DataFrame(index = excess_returns_3.columns, columns =
                           ['tangency','tangency with TIPS dropped',
                            'tangency with TIPS adjusted','equally-weighted',
                            'risk-parity','regularized'])
w_t_without_TIPS.loc['TIP']=0
w_t_compare['tangency'] = tangent_weights
w_t_compare['tangency with TIPS dropped'] = w_t_without_TIPS
w_t_compare['tangency with TIPS adjusted'] = w_t_adj_TIPS
w_t_compare['equally-weighted'] = 1/len(excess_returns.columns)
w_t_compare['risk-parity'] = 1/excess_returns.var()
w_t_compare['regularized'] = return_reg_tangent_weights(excess_returns)
w_t_compare
```

```python
w_t_3 = w_t_compare*target_mean/(excess_returns.mean()@w_t_compare)
w_t_3
```

```python
excess_returns_1 = excess_returns[excess_returns.columns[1:]]
w_t_result = excess_returns_1@w_t_3
w_t_stats = pd.DataFrame(index=w_t_result.columns)
w_t_stats['mean'] = w_t_result.mean()*12
w_t_stats['volatility'] = w_t_result.std()*np.sqrt(12)
w_t_stats['Sharpe ratio'] = (w_t_stats['mean']/w_t_stats['volatility'])*np.sqrt(12)
w_t_stats
```
